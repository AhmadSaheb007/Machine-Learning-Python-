```python
from sklearn.datasets import load_iris
import pandas as pd
#loading dataset . . . . . .
iris = load_iris()
# printing all datas . . . . . 
# print(iris.data)
iris_df = pd.DataFrame(iris.data, columns= ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'])
iris_df.head()

# adding spicies column . . . . . .
iris_df['spicies'] = iris.target
iris_df.head()
```

```python
# replacing spicies column with number 0, 1, 2
iris_df['spicies'] = iris_df['spicies'].replace(0, 'Setosa')
iris_df['spicies'] = iris_df['spicies'].replace(1, 'Verginidca')
iris_df['spicies'] = iris_df['spicies'].replace(2, 'Versicolor')
iris_df
```

# 2-D Scatter Plot using seaborn

```python
# 2-D Scatter plot with color-coding for each flower type/class.
# Here 'sns' corresponds to seaborn. 

import seaborn as sns
sns.set_style("whitegrid");
sns.FacetGrid(iris, hue="species", height=4) \
    .map(plt.scatter, "sepal_length", "sepal_width") \
    .add_legend();
plt.show();

# Notice that the blue points can be easily seperated 
# from red and green by drawing a line. 
# But red and green data points cannot be easily seperated.
# Can we draw multiple 2-D scatter plots for each combination of features?
# How many cobinations exist? 4C2 = 6.
```

# 3-D pair plot

```python
import plotly.express as px

df = px.data.iris()
fig = px.scatter_3d(df, x='sepal_length', y='sepal_width', z='petal_width',
              color='species')
fig.show()
```
