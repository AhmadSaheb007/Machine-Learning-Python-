```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#loading dataset . . . . . .
# loading the dataset
titanicDB = pd.read_csv('titanic.csv')

titanic_df = pd.DataFrame(titanicDB)
titanic_df.head()
```

# Replacing 0,1 to dead alive accordingly

```python
titanic_df['Survived'] = titanic_df['Survived'].replace(0, 'Dead')
titanic_df['Survived'] = titanic_df['Survived'].replace(1, 'Alive')
titanic_df
```

# Histogram

```python
sns.FacetGrid(titanicDB,hue="Survived",height=5)    \
    .map(sns.histplot,"Age")
plt.title("Histogram for Age")
plt.xlabel('AGE')
plt.ylabel('Survived')
plt.legend()
plt.show();

# Observation:
# People between 20-30 died and alive thewe most
# People greater then 50 yeasr died less but alive also less
```
