# Matplotlib

```python
from matplotlib import pyplot as plt
```

# Creating a besic graph

```python
X = list(range(1, 100, 10))
Y = list(range(1, 50, 5))
print(X)
print(Y)
plt.plot(X, Y)
plt.xlabel('Range 1-100')
plt.ylabel('Range 1-100')
plt.title('Range 1-100')
plt.show()  
```

# Plotting for non linear points

```python
x = [3, 4, 6, 10, 14]
y = [4, 6, 7, 5, 4]
z = [6, 8, 9, 10, 12]
plt.xlabel('Range 1-14')
plt.ylabel('Range 1-10')
plt.title('non linear plot')
plt.plot(x, y, color="red")
plt.plot(x, z, color="blue")
plt.legend(['RED', 'Blue'])
plt.show()
```

# Show dotted line in graph

```python
x = [3, 4, 5, 1, 8]
y = [1, 2, 3, 4, 5]
z = [3, 5, 9, 6, 2]
plt.scatter(x, y, color = "green")
plt.scatter(x, z, color = "red")
plt.show()
```

```python
x = [3, 4, 6, 10, 14]
y = [4, 6, 7, 5, 4]
z = [6, 8, 9, 10, 12]
plt.xlabel('Range 1-14')
plt.ylabel('Range 1-10')
plt.title('non linear plot')
plt.scatter(x, y, color="red")
plt.scatter(x, z, color="blue")
plt.legend(['RED', 'Blue'])
plt.grid() # adding grid
plt.show()
```

```python
x = ['Mango', 'Banana', 'Orange']
y = [5, 15, 10]
plt.bar(x, y)
plt.show()
```

# Designing the line plot

```python
X = list(range(1, 100, 10))
Y = list(range(1, 50, 5))
plt.plot(X, Y, linestyle='--', marker='o', color='red', linewidth=1, alpha=0.5)
plt.xlabel('Range 1-100')
plt.ylabel('Range 1-100')
plt.title('Range 1-100')
plt.show()  
```

# Making subplots

```python
x = [3, 4, 6, 10, 14]
y = [4, 6, 7, 5, 4]
z = [6, 8, 9, 10, 12]

# for x,y points . . . . . . . 
plt.subplot(2, 2, 1)
plt.title('x,y points non linear plot')
plt.plot(x, y, color="red")
plt.xlabel('Range 1-14')
plt.ylabel('Range 1-10')

# for x,z points . . . . . . . 
plt.subplot(2, 2, 2)
plt.plot(x, z, color="blue")
plt.xlabel('Range 1-14')
plt.ylabel('Range 1-10')
plt.title('x,z points non linear plot')
plt.legend()
plt.show()
```

```python
x = [3, 4, 6, 10, 14]
y = [4, 6, 7, 5, 4]
z = [6, 8, 9, 10, 12]
w = [2, 5, 9, 14, 20]

# for x,y points . . . . . . . 
plt.subplot(3, 1, 1)
plt.title('x,y points non linear plot')
plt.plot(x, y, color="red")
plt.xlabel('Range 1-14')
plt.ylabel('Range 1-10')
plt.show()

# for x,z points . . . . . . . 
plt.subplot(3, 1, 2)
plt.plot(x, z, color="blue")
plt.xlabel('Range 1-14')
plt.ylabel('Range 1-10')
plt.title('x,z points non linear plot')
plt.show()

# for x,w points . . . . . . . 
plt.subplot(3, 1, 3)
plt.plot(x, w, color="blue")
plt.xlabel('Range 1-14')
plt.ylabel('Range 1-10')
plt.title('x,z points non linear plot')
plt.show()
```

# making pie chart

```python
label = ['Mango', 'Banana', 'Orange', 'Chocolate', 'pizza']
size = [5, 15, 10, 50, 5]
plt.pie(size, labels = label, autopct = '%1.1f%%', shadow=True, startangle=90, explode=(0.3, 0.1, 0, 0.1, 0.2))
plt.show()
```
